```python
def haversine(lat1, lon1, lat2, lon2):
    # Calculate the distance between two points
    R = 6371  # Radius of the Earth in kilometers
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2) * np.sin(dlat / 2) + np.cos(np.radians(lat1)) \
        * np.cos(np.radians(lat2)) * np.sin(dlon / 2) * np.sin(dlon / 2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    d = R * c
    return d
```

```python
data_train['distance'] = haversine(data_train['Restaurant_latitude'], data_train['Restaurant_longitude'], data_train['Delivery_location_latitude'], data_train['Delivery_location_longitude'])
```

```python
y=data_train['Time_taken(min)']
```

```python
x=data_train[['Restaurant_latitude', 'Restaurant_longitude',
'Delivery_location_latitude', 'Delivery_location_longitude','Weatherconditions', 'Road_traffic_density', 'Festival','distance']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
pred_lr=lr.predict(x_test)
r2 = r2_score(y_test, pred_lr)

print('R-squared:', r2)
print("MSE with Linear Regression is :",mean_squared_error(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
pred_dt =dt.predict(x_test)

print("MSE with decision tree regressor:",mean_squared_error(y_test, pred_dt))  
r2_dtr = r2_score(y_test, pred_dt)
print('R-squared:', r2_dtr)
```

```python
from sklearn.linear_model import ElasticNet

# Create an instance of Elastic Net regression model
model = ElasticNet(alpha=0.05, l1_ratio=0.5)  # change alpha and l1_ratio as per your need

# Fit the Elastic Net model to the training data
model.fit(x_train, y_train)

#Predict the time taken on the testing data
pred_elastic  = model.predict(x_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, pred_elastic )
r2 = r2_score(y_test, pred_elastic )

print('MSE:', mse)
print('R2 score:', r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor().fit(x_train,y_train)
pred_rfr=rfr.predict(x_test)
print("MSE with Random forest regression:",mean_squared_error(y_test,pred_rfr))
r2_rfr = r2_score(y_test, pred_rfr)
print('R-squared:', r2_rfr)
```
